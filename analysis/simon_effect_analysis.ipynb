# Perfectionism and the Simon Effect

## 1. Data and Preprocessing

```python
import pandas as pd
df = pd.read_csv('Cognitie en Perceptie_data_schoon_2.csv', sep=';')
df['Q4'] = df['Q4'].map({'Man': 0, 'Vrouw': 1})
df = df.drop(index=0).reset_index(drop=True)
df = df.drop('Q6', axis=1)
df = df.dropna()
df
```

```python
import matplotlib.pyplot as plt
df = df.apply(pd.to_numeric, errors='coerce')
df.hist(bins=10, edgecolor='black', figsize=(12, 8))
plt.tight_layout()
plt.show()
```

## 2. Statistical Analysis

```python
from scipy.stats import shapiro
for col in df.columns:
    data = df[col].dropna()
    if len(data) >= 3: 
        stat, p = shapiro(data)
        print(f"{col}: p = {p:.5f}  {'Normaal verdeeld' if p > 0.05 else 'Niet normaal verdeeld'}")
```

```python
x = df['Q21_1']
y = df['Q12_3']

# Stap 4: Maak de scatterplot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, alpha=0.7)
plt.title('SAPS Standards vs. Simon Task Simon Effect')
plt.xlabel('SAPS Standards')
plt.ylabel('Simon Task/ Simon Effect (ms)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
x = df['Q21_2']
y = df['Q12_3']

# Stap 4: Maak de scatterplot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, alpha=0.7)
plt.title('SAPS Discrepancy vs. Simon Task Simon Effect')
plt.xlabel('SAPS Discrepancy')
plt.ylabel('Simon Task/ Simon Effect (ms)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
corr = df.corr(method='spearman') 

# 4. Heatmap tekenen
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlatie Heatmap')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import spearmanr
# Kolommen correct en numeriek maken
df['Q12_3'] = pd.to_numeric(df['Q12_3'], errors='coerce')
df['Q21_1'] = pd.to_numeric(df['Q21_1'], errors='coerce')
df['Q21_2'] = pd.to_numeric(df['Q21_2'], errors='coerce')

# Verwijder missende waarden
data = df[['Q12_3', 'Q21_1', 'Q21_2']].dropna()

# Pearson-correlaties en p-waarden
r_std, p_std = spearmanr(data['Q12_3'], data['Q21_1'])
r_disc, p_disc = spearmanr(data['Q12_3'], data['Q21_2'])

print(f"Simon Effect vs SAPS Standards: r = {r_std:.2f}, p = {p_std:.4f}")
print(f"Simon Effect vs SAPS Discrepancy: r = {r_disc:.2f}, p = {p_disc:.4f}")
```

```python
import numpy as np# Scatterplot
plt.figure(figsize=(10, 6))

# Standards
plt.scatter(df['Q12_3'], df['Q21_1'], color='blue', label='SAPS Standards')
z1 = np.polyfit(df['Q12_3'], df['Q21_1'], 1)
plt.plot(df['Q12_3'], np.polyval(z1, df['Q12_3']), color='blue')

# Discrepancy
plt.scatter(df['Q12_3'], df['Q21_2'], color='orange', label='SAPS Discrepancy')
z2 = np.polyfit(df['Q12_3'], df['Q21_2'], 1)
plt.plot(df['Q12_3'], np.polyval(z2, df['Q12_3']), color='orange')

# Labels
plt.title('Verband tussen Simon Effect en SAPS-scores')
plt.xlabel('Simon Effect')
plt.ylabel('SAPS Score')
plt.legend()
plt.tight_layout()
plt.show()
```
